```python
# only in titanic_df, fill the two missing values with the most occurred value, which is "S".
titanic_main["Embarked"] = titanic_main["Embarked"].fillna("S")

# plot
sns.factorplot('Embarked','Survived', data=titanic_main,size=4,aspect=3)

fig, (axis1,axis2,axis3) = plt.subplots(1,3,figsize=(15,5))

sns.countplot(x='Embarked', data=titanic_main, ax=axis1)
sns.countplot(x='Survived', hue="Embarked", data=titanic_main, order=[1,0], ax=axis2)

embark_perc = titanic_main[["Embarked", "Survived"]].groupby(['Embarked'],as_index=False).mean()
sns.barplot(x='Embarked', y='Survived', data=embark_perc,order=['S','C','Q'],ax=axis3)
```
